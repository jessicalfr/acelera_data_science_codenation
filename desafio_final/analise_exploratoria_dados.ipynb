### Base de dados

A principal base de dados do projeto é a que foi chamada de `dados_mercado`. Ela contém as informações de todas as empresas que estão no mercado analisado. O objetivo desse notebook é fazer uma análise descritiva dessa base e então iniciar a construção da solução.

```python
# leitura da base de dados
dados_mercado = pd.read_csv('./data/estaticos_market.csv')
dados_mercado.shape
```

A base contém 182 colunas com dados de 462.298 empresas. A primeira coisa verificada será os tipos das variáveis e a presença de NAs.

```python
# metadados
metadados = pd.DataFrame({'Tipo': dados_mercado.dtypes,
                          'NAs': dados_mercado.isna().sum()})
metadados.index.name = 'Coluna'
metadados.reset_index(inplace=True)

metadados
```

```python
# número de variáveis sem NAs
metadados[metadados.NAs==0].shape
```

A base tem muitas colunas com muitos NAs. Apenas 17 atributos do banco de dados não têm NAs (removendo as colunas id e Unnamed: 0).

```python
# menos de 10% de NAs
metadados_possiveis = metadados[metadados.NAs <= dados_mercado.shape[0]*0.10]
metadados_possiveis
```

Apenas serão utilizadas as colunas com NA em até 10% dos registros. Isso será feito pois colunas com grande proporção de NAs tendem a ser pouco informativas para o todo. A estratégia adotada é a seguinte:

- Nas colunas de tipo `object`, os NAs serão substituídos pela categoria `SEM INFORMACAO`.
- Nas colunas de tipo `float64` e `int64`, os NAs serão substituídos pela média das observações.
- Nas colunas de tipo `bool`, os NAs serão substituídos pelo valor `False`.

Além disso, serão removidas 8 colunas pelos seguintes motivos:

- `de_ramo`: por se tratar de uma descrição.
- `de_natureza_juridica`: pois existe outra coluna no banco que tem a mesma informação de forma macro (`natureza_juridica_macro`)
- `nm_divisao`: por conter muitas categorias (é uma coluna quase descritiva).
- `idade_emp_cat`: por conter a mesma informação de `idade_empresa_anos`, porém categorizada.
- `dt_situacao`: por ser uma variável de data ligada a uma variável descritiva (`de_situacao`).
- `de_saude_rescencia`: por ter a mesma informação contida em `nu_meses_rescencia`, categorizada.
- `de_faixa_faturamento_estimado`: por ter a mesma informação contida em `vl_faixa_faturamento_estimado_aux`.
- `de_faixa_faturamento_estimado_grupo`: por ter a mesma informação contida em `vl_faixa_faturamento_estimado_grupo_aux`.

Por fim, as variáveis `fl_spa`, `fl_anti`, `fl_veiculo`, `fl_simples_irregular` e `fl_possivel_iss` terão seu tipo corrigido para `bool`.

```python
# separa as colunas que serão utilizadas
colunas_utilizadas = list(metadados_possiveis['Coluna'][1:])

cols_remover = ['de_ramo','de_natureza_juridica','nm_divisao','idade_emp_cat','dt_situacao','de_saude_rescencia',
                'de_faixa_faturamento_estimado','de_faixa_faturamento_estimado_grupo']

for col in cols_remover:
    colunas_utilizadas.remove(col)

len(colunas_utilizadas)
```

```python
# corrige o tipo das colunas
colunas_bool = ['fl_spa','fl_antt','fl_veiculo','fl_simples_irregular','fl_passivel_iss']

for col in colunas_bool:
    dados_mercado[col] = dados_mercado[col].astype('bool')
```

```python
# separa as colunas relevantes
dados_mercado_prep = dados_mercado.loc[:, colunas_utilizadas]

# faz a imputação dos dados
# cria metadados das colunas
metadados_util = pd.DataFrame({'Tipo': dados_mercado_prep.dtypes,
                               'NAs': dados_mercado_prep.isna().sum()})
metadados_util.index.name = 'Coluna'
metadados_util.reset_index(inplace=True)

# categóricos
for col in metadados_util.loc[metadados_util.Tipo=='object', 'Coluna']:
    dados_mercado_prep[col].fillna('SEM INFORMACAO', inplace = True)
            
# inteiros
for col in metadados_util.loc[metadados_util.Tipo=='int64', 'Coluna']:
    media_int = dados_mercado_prep[col].mean()
    dados_mercado_prep[col].fillna(media_int, inplace = True)
    
# float
for col in metadados_util.loc[metadados_util.Tipo=='float64', 'Coluna']:
    media_flt = dados_mercado_prep[col].mean()
    dados_mercado_prep[col].fillna(media_flt, inplace = True)

# booleanos
for col in metadados_util.loc[metadados_util.Tipo=='bool', 'Coluna']:
    dados_mercado_prep[col].fillna(False, inplace = True)
```

```python
# verifica o resultado
dados_mercado_prep.info()
```

Com isso, foram verificadas todas as variáveis que foram pre-selecionadas, restando 28 variáveis que serão utilizadas no sistema de recomendação.

Por último, é necessário fazer *one-hot encoding* das variáveis categóricas e padronizar as variáveis numéricas.

```python
# cria metadados final
metadados_final = pd.DataFrame({'Tipo': dados_mercado_prep.dtypes})
metadados_final.index.name = 'Coluna'
metadados_final.reset_index(inplace=True)
```

```python
# remove a coluna id das variáveis afetadas
cols_one_hot = metadados_final.loc[metadados_final.Tipo=='object', 'Coluna']
cols_one_hot = list(cols_one_hot)
cols_one_hot.remove('id')

# one-hot encoding
dados_mercado_final = pd.get_dummies(dados_mercado_prep.loc[:, cols_one_hot])

# adiciona a coluna de id
dados_mercado_final.insert(0, 'id', dados_mercado_prep['id'])
```

```python
# visualiza as colunas
dados_mercado_final.head()
```

```python
# visualiza o shape
dados_mercado_final.shape
```

```python
# cria lista de variáveis numéricas
cols_numericas = (metadados_final['Tipo']=='int64') | (metadados_final['Tipo']=='float64')
cols_normaliza = metadados_final.loc[cols_numericas, 'Coluna']
cols_normaliza = list(cols_normaliza)

# normaliza as variáveis numéricas
min_max_scaler = preprocessing.MinMaxScaler()

for col in cols_normaliza:
    dados_mercado_final[col] = min_max_scaler.fit_transform(dados_mercado_prep[[col]])
```

```python
# visualiza as colunas
dados_mercado_final.head()
```

```python
# visualiza o shape
dados_mercado_final.shape
```

```python
# adiciona as variáveis binárias
cols_bool = metadados_final.loc[metadados_final.Tipo=='bool', 'Coluna']
cols_bool = list(cols_bool)

# transforma em 0 ou 1 
dados_mercado_final[cols_bool] = dados_mercado_prep[cols_bool].astype('int64')
```
